```python
import numpy as np
from matplotlib import pyplot as plt

def taylor_sin(x, terms=9):
    return np.polyval([(1 if i%4==1 else -1)/np.math.factorial(i)  if (i%2) else 0 for i in range(terms, -1, -1)], x)
    return sum(-1**n for n in range(0,terms))

def taylor_cos(x, terms=9):
    return np.polyval([(1 if i%4==0 else -1)/np.math.factorial(i)  if (i%2)==0 else 0 for i in range(terms, -1, -1)], x)
    return sum(-1**n for n in range(0,terms))

plt.xlim(-10, 10)
plt.ylim(-20, 20)
plt.grid(True)

x=np.linspace(-5, 5, 1000)
plt.plot(x, np.tan(x), 'k', linewidth=2)
y_cos = taylor_cos(x)
y_sin = taylor_sin(x)
y_tan = y_sin/y_cos

plt.plot(x, y_tan, 'g--', linewidth=2)

```
