```python
import numpy as np
from matplotlib import pyplot as plt

A = np.array([[0.5, 0.5, 0.25],
              [0.25, 0.0, 0.25],
              [0.25, 0.5, 0.5]])
print(A)
xtoday = np.array([[0],[1],[0]])
weather=np.zeros((3,25))
for k in range(weather.shape[1]):
    weather[:,k]=xtoday[:,0]
    xtomorrow = A@xtoday
    xtoday = xtomorrow
    # print(f"{k}:\n{xtoday}\n")
print(weather)
plt.plot(weather.transpose())
plt.grid(True)
```
