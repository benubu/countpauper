```python
# https://www.youtube.com/watch?v=ebfOSDj4j3I&list=PLMrJAkhIeNNTYaOnVI3QpH7jgULnAmvPA&index=6

import numpy as np 
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp
from scipy.special import factorial 

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True)

# basic sin
x=np.linspace(-10, 10, 2000)
y = np.sin(x)
plt.plot(x,y, 'k', linewidth=2)

# first order 
P = [1, 0] # x + 0
yT1 = np.polyval(P,x)
plt.plot(x, yT1, 'b--', linewidth = 1.2)

# 3rd order 
P = [-1 / factorial(3), 0, 1, 0] # x + 0
yT3 = np.polyval(P,x)
plt.plot(x, yT3, 'r--', linewidth = 1.2)

# 5th order 
P = [1 / factorial(5), 0, -1 / factorial(3), 0, 1, 0] # x + 0
yT5 = np.polyval(P,x)
plt.plot(x, yT5, 'g--', linewidth = 1.2)

#7th order 
P = [-1/factorial(7), 0, 1 / factorial(5), 0, -1 / factorial(3), 0, 1, 0] # x + 0
yT7 = np.polyval(P,x)
plt.plot(x, yT7, 'm--', linewidth = 1.2)

# final
order = 15
P = [(1 if i%4==1 else -1)/factorial(i)  if (i%2) else 0 for i in range(order, -1, -1)]
yT = np.polyval(P,x)
plt.plot(x, yT, 'c--', linewidth = 1.2)


plt.legend(['sin(x)', 'linear', 'cubic', 'quintic', 'septic', 'final'])
plt.show()
```
