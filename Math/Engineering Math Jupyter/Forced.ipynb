```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
from scipy import signal

A=np.array([[0,1],[-1,-0.1]])
B=np.array([[0],[1]])
C=np.eye(2)
D=np.array([[0],[0]])

sys=signal.StateSpace(A,B,C,D)

t,y=signal.impulse(sys,100)
plt.plot(t,y)
plt.grid(True)
plt.show()
```
